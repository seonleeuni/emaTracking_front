```python
import os
import pandas as pd
from datetime import datetime

path = 'data/samsunghealth_821074645649_20250121171455/'
entries = os.listdir('data/samsunghealth_821074645649_20250121171455')
data_name = 'com.samsung.shealth.tracker.pedometer_step_count.20250121171455.csv'
temp = pd.read_csv(path+data_name, skiprows=1,nrows=1, header=None)
col_num = temp.columns[-1]


######################## phone data 종류


# Walk step
#filename = com.samsung.shealth.tracker.pedometer_step_count.(user_num).csv
#com.samsung.health.step_count.start_time : 재기 시작
#com.samsung.health.step_count.end_time : 재기 끝남
#duration : 총 잰 시간(ms)
#com.samsung.health.step_count.count : 걸음수
#walk_step, run_step : 걸은 걸음수, 뛴 걸음 수
# com.samsung.health.step_count.speed : 스피드 (m/s)
#com.samsung.health.step_count.distance : 움직임 (m)
#com.samsung.health.step_count.calorie : calorie

cols_original = {'walk':['com.samsung.health.step_count.start_time','com.samsung.health.step_count.end_time','duration','run_step','walk_step','com.samsung.health.step_count.count','com.samsung.health.step_count.speed','com.samsung.health.step_count.distance','com.samsung.health.step_count.calorie'],
                'excercise':['com.samsung.health.exercise.start_time','com.samsung.health.exercise.end_time','com.samsung.health.exercise.duration','com.samsung.health.exercise.exercise_type','com.samsung.health.exercise.mean_heart_rate', 'com.samsung.health.exercise.max_heart_rate','com.samsung.health.exercise.min_heart_rate','com.samsung.health.exercise.max_speed','com.samsung.health.exercise.mean_speed','com.samsung.health.exercise.distance','com.samsung.health.exercise.count_type','com.samsung.health.exercise.count', 'com.samsung.health.exercise.calorie','com.samsung.health.exercise.exercise_custom_type','com.samsung.health.exercise.location_data','com.samsung.health.exercise.live_data'],
                'heart_rate':[]}

cols_var = {'walk': ['start','end','duration','run_step','walk_step','total_step','speed','distance','calorie'],
           'excercise':['start','end','duration','exercise_type','mean_heart_rate', 'max_heart_rate','min_heart_rate','max_speed','mean_speed','total_distance','count_type','count', 'calorie','exercise_custom_type','location_data','live_data'],
            'heart_rate':
           }






class DataProcessor:
    def __init__(self, path):
        self.path = path
        
    def data_extract(self,data_name):
        """
        CSV 파일을 읽고, 마지막 컬럼의 개수를 확인한 후 데이터를 반환
        """
        temp = pd.read_csv(self.path + data_name, skiprows=1, nrows=1, header=None)
       
        col_num = temp.columns[-1]+1  # 총 column 갯수
        print(col_num)
        
        data = pd.read_csv(self.path + data_name, header=1, index_col=col_num)
        return data
        
    def extract_period(self, data, start_var,date_start, date_end):
        """
        주어진 기간(date_start ~ date_end)에 해당하는 데이터를 필터링
        """
        data.reset_index(drop=True,inplace=True)
        print(data.shape)
        
        data[start_var] = pd.to_datetime(data[start_var], format="%Y-%m-%d %H:%M:%S.%f",errors="coerce")
        
        start = datetime.strptime(date_start, "%Y-%m-%d")
        end = datetime.strptime(date_end, "%Y-%m-%d")
        
        data_new = data[(data[start_var] < end) & (data[start_var] > start)]
        return data_new
```

```python
processor = DataProcessor(path)
data_name = 'com.samsung.shealth.tracker.pedometer_step_count.20250121171455.csv'
data_walk = processor.data_extract(data_name)
processor.extract_period(data_walk, "com.samsung.health.step_count.start_time","2024-12-01","2024-12-31")
```
